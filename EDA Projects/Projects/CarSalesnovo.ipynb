```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Ciência de dados /Projeto 1 /Car_sales.csv')
```

TRATAMENTO DOS DADOS

A) RENOMEAR TODAS AS COLUNAS PARA PORTUGUES

```python
novos_nomes = {
    'Manufacturer':'Fabrica',
    'Model': 'Modelo',
    'Sales_in_thousands': 'Vendas_em_milhares',
    '__year_resale_value' : 'Valor_de_revenda_apos_um_ano',
    'Vehicle_type' : 'Tipo_de_veiculo',
    'Price_in_thousands': 'Preco_em_milhares',
    'Engine_size': 'Tamanho_do_motor',
    'Horsepower': 'Potencia_em_cavalos_de_forca',
    'Wheelbase': 'Distancia_entre_os_eixos',
    'Width': 'Largura',
    'Length': 'Comprimento',
    'Curb_weight': 'Peso_do_carro',
    'Fuel_capacity': 'Capacidade_do_tanque_de_combustivel',
    'Fuel_efficiency': 'Eficiencia_de_combustivel',
    'Latest_Launch': 'Ultimo_lancamento',
    'Power_perf_factor': 'Fator_de_desempenho_de_potencia'
}
df.rename(columns = novos_nomes, inplace=True)
```

B) VERIFIQUEI A QUANTIDADE DE VALORES NULOS E SUBSTITUI PELA MEDIA

```python
mediana_1 = df['Valor_de_revenda_apos_um_ano'].mean()
df['Valor_de_revenda_apos_um_ano'] = df['Valor_de_revenda_apos_um_ano'].fillna(mediana_1)
mediana_2 = df['Preco_em_milhares'].mean()
mediana_3 = df['Tamanho_do_motor'].mean()
mediana_4 = df['Potencia_em_cavalos_de_forca'].mean()
mediana_5 = df['Distancia_entre_os_eixos'].mean()
mediana_6 = df['Largura'].mean()
mediana_7 = df['Comprimento'].mean()
mediana_8 = df['Peso_do_carro'].mean()
mediana_9 = df['Capacidade_do_tanque_de_combustivel'].mean()
mediana_10 = df['Eficiencia_de_combustivel'].mean()
mediana_11 = df['Fator_de_desempenho_de_potencia'].mean()
df['Preco_em_milhares']= df['Preco_em_milhares'].fillna(mediana_2)
df['Tamanho_do_motor']=df['Tamanho_do_motor'].fillna(mediana_3)
df['Potencia_em_cavalos_de_forca']=df['Potencia_em_cavalos_de_forca'].fillna(mediana_4)
df['Distancia_entre_os_eixos']=df['Distancia_entre_os_eixos'].fillna(mediana_5)
df['Largura']=df['Largura'].fillna(mediana_6)
df['Comprimento']=df['Comprimento'].fillna(mediana_7)
df['Peso_do_carro']=df['Peso_do_carro'].fillna(mediana_8)
df['Capacidade_do_tanque_de_combustivel']=df['Capacidade_do_tanque_de_combustivel'].fillna(mediana_9)
df['Eficiencia_de_combustivel']=df['Eficiencia_de_combustivel'].fillna(mediana_10)
df['Fator_de_desempenho_de_potencia']=df['Fator_de_desempenho_de_potencia'].fillna(mediana_11)
df.isnull().sum()
```

Qual é a relação entre o preço do carro e suas vendas?

```python
correlacao = df['Preco_em_milhares'].corr(df['Vendas_em_milhares'])
print(correlacao)
# o resultado foi próximo de -1 o que indica que preços mais altos tem relação com vendas mais baixas, ou seja quanto maior o precomenor sera a vendas, no entanto o numero do resultado deu mais próximode zero doque -1, o que indica também a ocorrencia de nao ter correlação linear, usarei machine leaning para verificar essa resposta
x = df['Preco_em_milhares']
y = df['Vendas_em_milhares']
plt.rcParams['figure.figsize'] =[10,6]
plt.scatter(x, y)
plt.grid(True)
plt.show()
plt.figure(figsize=(10, 6))
sns.regplot(x='Preco_em_milhares', y='Vendas_em_milhares', data=df)
plt.xlabel('Preço do Carro (milhares de dólares)')
plt.ylabel('Vendas do Carro (milhares de unidades)')
plt.title('Relação entre Preço do Carro e Vendas')
plt.grid(True)  # Adicionando grade ao gráfico
plt.show()

```
